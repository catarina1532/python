```python
import pandas as pd

data = {'Product': ['A', 'B', 'C', 'D', 'E', 'F'],
        'Price': [10, 25, 15, 30, 20, 35],
        'Quantity': [100, 50, 75, 30, 90, 20],
        'Sales': [1000, 1250, 1125, 900, 1800, 700]}

df = pd.DataFrame(data)
df = df.set_index('Product') #use the column Product as the index (if i didn't write this line of code it would generate a column primary key)
display(df)
```

```python
import pandas as pd

data = {'Product': [x.lower() for x  in ['A', 'B', 'C', 'D', 'E', 'F']], #make the letters case insensitive instead of changing to ['a', 'b', 'c', 'd', 'e', 'f']
        'Price': [10, 25, 15, 30, 20, 35],
        'Quantity': [100, 50, 75, 30, 90, 20],
        'Sales': [1000, 1250, 1125, 900, 1800, 700]}

df = pd.DataFrame(data)
df = df.set_index('Product') #use the column Product as the index (if i didn't write this line of code it would generate a column primary key)
display(df)
```

```python
df['Price'] #if i just want the column Price (index is still there)
#type(df['Price']) #result is a series
#type(df[['Price']]) #result is a dataframe
```

```python
df['Sales'] > 1000 #result is a series with TRUE or FALSE
```

```python
df[df['Sales'] > 1000] #select from the dataframe which rows satisfy that condition
```

```python
#iloc (select from my table - extract a subtable)
#define which rows and columns i'm interested from my table
#index by inumerating them
display(df.iloc[0:2, :]) #the first two rows and all columns
```

```python
display(df.iloc[0:2, :1]) #only the first column of the two rows
```

```python
display(df.iloc[2, 1]) #extract a single value (returns a value in NumPy)
```

```python
display(df.iloc[1:4, 0:2]) #slice 1:4 starts at 1 and ends at 3
```
